# Top 10 Most Rented Films

```python
%%sql
SELECT film.title AS Title, COUNT(rental.rental_id) AS rental_count
FROM film 
JOIN inventory  ON film.film_id = inventory.film_id
JOIN rental  ON inventory.inventory_id = rental.inventory_id
GROUP BY film.title
ORDER BY rental_count DESC
LIMIT 10;
```

# Total Revenue per Customer

```python
%%sql
SELECT customer.customer_id, customer.first_name || ' ' || customer.last_name AS customer_name,
       SUM(payment.amount) AS total_revenue
FROM customer 
JOIN payment ON customer.customer_id = payment.customer_id
GROUP BY customer.customer_id, customer_name
ORDER BY total_revenue DESC;
```

# Top 3 Actors by Number of Films in Each Category

```python
%%sql

SELECT category, actor_name, film_count
FROM (
    SELECT category.name AS category,
           actor.first_name || ' ' || actor.last_name AS actor_name,
           COUNT(*) AS film_count,
           DENSE_RANK() OVER (PARTITION BY category.name ORDER BY COUNT(*) DESC) AS rank
    FROM film 
    JOIN film_actor ON film.film_id = film_actor.film_id
    JOIN actor ON actor.actor_id = film_actor.actor_id
    JOIN film_category ON film.film_id = film_category.film_id
    JOIN category ON category.category_id = film_category.category_id
    GROUP BY category.name, actor_name
) ranked_actors
WHERE rank <= 3;
```

# Most Profitable Film

```python
%%sql

SELECT film.title, SUM(payment.amount) AS total_revenue
FROM film 
JOIN inventory ON film.film_id = inventory.film_id
JOIN rental ON inventory.inventory_id = rental.inventory_id
JOIN payment ON rental.rental_id = payment.rental_id
GROUP BY film.title
ORDER BY total_revenue DESC
LIMIT 1;
```

# Monthly Revenue Trend

```python
%%sql 

SELECT DATE_TRUNC('month', payment_date) AS month,
       SUM(amount) AS monthly_revenue
FROM payment
GROUP BY month
ORDER BY month;
```

# Customers Who Never Made a Payment

```python
%%sql

SELECT customer.customer_id, customer.first_name, customer.last_name
FROM customer 
WHERE NOT EXISTS (
    SELECT 1 FROM payment WHERE payment.customer_id = customer.customer_id
);
```

# Overdue Rentals

```python
%%sql

SELECT rental.rental_id, customer.first_name || ' ' || customer.last_name AS customer_name,
       film.title, rental.rental_date
FROM rental 
JOIN inventory ON rental.inventory_id = inventory.inventory_id
JOIN film ON inventory.film_id = film.film_id
JOIN customer ON rental.customer_id = customer.customer_id
WHERE rental.return_date IS NULL;
```

# Number of Films by Category

```python
%%sql

SELECT category.name AS category, COUNT(film_category.film_id) AS film_count
FROM category 
JOIN film_category ON category.category_id = film_category.category_id
GROUP BY category.name
ORDER BY film_count DESC;
```

# Store Revenue and Top Performing Staff per Store

```python
%%sql 

WITH store_revenue AS (
  SELECT store.store_id, staff.staff_id,
         SUM(payment.amount) AS total_sales
  FROM store 
  JOIN staff ON store.store_id = staff.store_id
  JOIN payment ON staff.staff_id = payment.staff_id
  GROUP BY store.store_id, staff.staff_id
),
ranked_sales AS (
  SELECT *, RANK() OVER (PARTITION BY store_id ORDER BY total_sales DESC) AS rank
  FROM store_revenue
)
SELECT store_id, staff_id, total_sales
FROM ranked_sales
WHERE rank = 1;
```

# Films Never Rented

```python
%%sql

SELECT film.film_id, film.title
FROM film 
WHERE NOT EXISTS (
    SELECT 1
    FROM inventory 
    JOIN rental ON inventory.inventory_id = rental.inventory_id
    WHERE inventory.film_id = film.film_id
);
```
